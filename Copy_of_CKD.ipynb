```python
def draw_histograms(df):
    numeric_columns = df.select_dtypes(include='number')
    for col in numeric_columns.columns:
        plt.figure(figsize=(6, 4))
        sns.histplot(df[col], bins=30, kde=True)
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('count')
        plt.tight_layout()
        plt.show()
draw_histograms(df)
```

```python
def draw_boxplots(df):
    numeric_columns = df.select_dtypes(include='number')
    for col in numeric_columns.columns:
        plt.figure(figsize=(6, 5))
        sns.boxplot(df,y=col)
        plt.title(f'Boxplot of {col}')
        plt.ylabel(col)
        plt.tight_layout()
        plt.show()
draw_boxplots(df)
```

```python
def plot_categorical_counts(df):
    count_columns = df.select_dtypes(include='object').columns
    for col in count_columns:
        plt.figure(figsize=(8, 4))
        sns.countplot(data=df, x=col,hue=df['Target'])
        plt.title(f'Count Plot of {col}')
        plt.tight_layout()
        plt.show()
plot_categorical_counts(df)
```

```python
def get_outliers_iqr(df):
    numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
    outlier_data = []

    for col in numeric_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
        for idx, val in outliers.items():
            outlier_data.append({'Column': col, 'Index': idx, 'Outlier Value': val})

    return pd.DataFrame(outlier_data)


outliers_df = get_outliers_iqr(df)
print(outliers_df)
```

```python
from sklearn.preprocessing import LabelEncoder
def label_encode_object_columns(df, exclude_columns=None, save_path='label_encoders'):
    df_encoded = df.copy()
    label_encoders = {}

    object_columns = df_encoded.select_dtypes(include='object').columns


    if exclude_columns:
        object_columns = [col for col in object_columns if col not in exclude_columns]

    for col in object_columns:
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
        label_encoders[col] = le



    return df_encoded, label_encoders
df_encoded,label_encoder  = label_encode_object_columns(df,exclude_columns=['Target'])
```

```python
from sklearn.preprocessing import StandardScaler

def apply_standard_scaler(df, discrete_columns):
    df_scaled = df.copy()


    numeric_columns = df.select_dtypes(include='number').columns


    columns_to_scale = [col for col in numeric_columns if col not in discrete_columns]


    scaler = StandardScaler()
    df_scaled[columns_to_scale] = scaler.fit_transform(df_scaled[columns_to_scale])

    return df_scaled, scaler, columns_to_scale
df_scaled, scaler, scaled_columns = apply_standard_scaler(df_encoded, discrete_columns)
```

```python
from scipy.stats import chi2_contingency

alpha = 0.1


cat_features = [
    'Albumin in urine', 'Sugar in urine',
    'Red blood cells in urine', 'Pus cells in urine',
    'Pus cell clumps in urine', 'Bacteria in urine',
    'Hypertension (yes/no)', 'Diabetes mellitus (yes/no)',
    'Coronary artery disease (yes/no)', 'Appetite (good/poor)',
    'Pedal edema (yes/no)', 'Anemia (yes/no)',
    'Family history of chronic kidney disease', 'Smoking status',
    'Physical activity level', 'Urinary sediment microscopy results'
]


chi2_results = {}

for col in cat_features:
    contingency = pd.crosstab(df_scaled[col], df_encoded['Target'])
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }


chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')

chi2_df
```

```python
from sklearn.feature_selection import mutual_info_classif, f_classif



X = df_encoded.drop(columns=['Target'])
y = df_encoded['Target']


mi_scores = mutual_info_classif(X, y, discrete_features='auto', random_state=42)
mi_scores = pd.Series(mi_scores, index=X.columns).sort_values(ascending=False)


f_scores, p_values = f_classif(X, y)
f_scores = pd.Series(f_scores, index=X.columns).sort_values(ascending=False)
p_values = pd.Series(p_values, index=X.columns).sort_values()


print("Top features by Mutual Information:")
print(mi_scores.head(10))


print("\nTop features by ANOVA F-test:")
print(f_scores.head(10))


feature_scores = pd.DataFrame({
    'Mutual Information': mi_scores,
    'ANOVA F-score': f_scores,
    'ANOVA p-value': p_values
}).sort_values(by='Mutual Information', ascending=False)

print("\nCombined feature scores:")
print(feature_scores.head(10))
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def cramers_v(x, y):
    contingency_table = pd.crosstab(x, y)
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    n = contingency_table.sum().sum()
    phi2 = chi2 / n
    r, k = contingency_table.shape
    phi2_corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    r_corr = r - ((r-1)**2)/(n-1)
    k_corr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2_corr / min((k_corr-1), (r_corr-1)))



cat_features = [
    'Albumin in urine', 'Sugar in urine', 'Red blood cells in urine', 'Pus cells in urine',
    'Pus cell clumps in urine', 'Bacteria in urine', 'Hypertension (yes/no)',
    'Diabetes mellitus (yes/no)', 'Coronary artery disease (yes/no)', 'Appetite (good/poor)',
    'Pedal edema (yes/no)', 'Anemia (yes/no)', 'Family history of chronic kidney disease',
    'Smoking status', 'Physical activity level', 'Urinary sediment microscopy results'
]

cramers_results = {}

for col in cat_features:
    cramers_results[col] = cramers_v(df_encoded[col], df_encoded['Target'])


cramers_df = pd.DataFrame.from_dict(cramers_results, orient='index', columns=['Cramers_V'])
cramers_df = cramers_df.sort_values(by='Cramers_V', ascending=False)

print(cramers_df)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X_scaled = df_scaled.drop(columns=['Target']) 
y = df_scaled['Target'] 


pca = PCA(n_components=27)
X_pca = pca.fit_transform(X_scaled)


pca_df = pd.DataFrame(X_pca[:, :2], columns=['PC1', 'PC2'])
pca_df['Target'] = y.values


plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Target', palette='Set2', s=80)
plt.title('PCA Projection (First Two Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42, stratify=y)


svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(X_train, y_train)


y_pred = svm_model.predict(X_test)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

print("Random Forest Results:")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
print("Accuracy Score:", accuracy_score(y_test, y_pred_rf))
```
