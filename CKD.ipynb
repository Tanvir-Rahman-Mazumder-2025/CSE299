```python
def draw_histograms(df):
    numeric_columns = df.select_dtypes(include='number')
    for col in numeric_columns.columns:
        plt.figure(figsize=(6, 4))
        sns.histplot(df[col], bins=30, kde=True)
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('count')
        plt.tight_layout()
        plt.show()
draw_histograms(df)
```

```python
def plot_categorical_counts(df):
    cat_cols = df.select_dtypes(include='object').columns
    for col in cat_cols:
        plt.figure(figsize=(8, 4))
        sns.countplot(data=df, x=col,hue=df['Target'])
        plt.title(f'Count Plot of {col}')
        plt.tight_layout()
        plt.show()
plot_categorical_counts(df)
```

```python
def draw_boxplot(df):
    numeric_columns = df.select_dtypes(include='number')
    for col in numeric_columns.columns:
        plt.figure(figsize=(6, 4))
        sns.boxplot(x=df["Target"],y=df[col] )
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('count')
        plt.tight_layout()
        plt.show()
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
        print(f"{col} → {len(outliers)} outliers found")

draw_boxplot(df)
```

```python
def plot_heatmap_batches(df, size=10, target_col='Target'):
    numeric_columns = df.select_dtypes(include='number').columns.tolist()
    if target_col in numeric_columns:
        numeric_columns.remove(target_col)
    total = len(numeric_columns)
    for i in range(0, total, size):
        batch = numeric_columns[i:i+size]
        cols = [target_col] + batch
        corr_matrix = df[cols].corr()
        plt.figure(figsize=(12, 8))
        sns.heatmap(corr_matrix, annot=True,  linewidths=0.5)
        plt.title(f'Correlation Heatmap with Target (Features {i+1} to {i+len(batch)})')
        plt.tight_layout()
        plt.show()
plot_heatmap_batches(df)
```

```python
correlations = df.corr(numeric_only=True)['Target'].drop('Target')
sorted_corr = correlations.abs().sort_values(ascending=False)
top_features = sorted_corr.head(15)
print(top_features)
```

```python
from sklearn.preprocessing import LabelEncoder

# Make a copy to avoid changing original data
df_encoded = df.copy()

# Apply label encoding to object columns
label_encoder = LabelEncoder()

for col in df_encoded.select_dtypes(include="object").columns:
    # Apply only if column has 2 unique values (binary)
    if df_encoded[col].nunique() == 2:
        df_encoded[col] = label_encoder.fit_transform(df_encoded[col])
```
