```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df_main = pd.read_csv(r'kidney_disease_dataset.csv')
```

```python
df_main['Target'].value_counts()
```

```python
df_noRisk = df_main[df_main['Target'] == 'No_Disease'].sample(n=410, random_state=42)
df_lowRisk = df_main[df_main['Target'] == 'Low_Risk'].sample(n=410, random_state=42)
df_Moderate_Risk = df_main[df_main['Target'] == 'Moderate_Risk'].sample(n=410, random_state=42)

df_High_Risk = df_main[df_main['Target'] == 'High_Risk'].sample(n=410, random_state=42)

df_Severe_Disease = df_main[df_main['Target'] == 'Severe_Disease'].sample(n=410, random_state=42)


df_equal = pd.concat([df_noRisk,df_lowRisk,df_Moderate_Risk,df_High_Risk,df_Severe_Disease], ignore_index=True)


df_equal = df_equal.sample(frac=1, random_state=44).reset_index(drop=True)


df_equal.to_csv('equal_data.csv', index=False)
```

```python
object_columns = [
    'Red blood cells in urine','Pus cells in urine','Pus cell clumps in urine','Bacteria in urine','Hypertension (yes/no)',
    'Diabetes mellitus (yes/no)','Coronary artery disease (yes/no)','Appetite (good/poor)','Pedal edema (yes/no)','Anemia (yes/no)',
    'Family history of chronic kidney disease','Smoking status','Physical activity level','Urinary sediment microscopy results'
]
from sklearn.preprocessing import LabelEncoder
def label_encode_object_columns(df_equal,save_path='label_encoders'):
    df_encoded = df_equal.copy()
    label_encoders = {}



    for col in object_columns:
        LE = LabelEncoder()
        df_encoded[col] = LE.fit_transform(df_encoded[col].astype(str))
        label_encoders[col] = LE



    return df_encoded, label_encoders
df_encoded,label_encoder  = label_encode_object_columns(df_equal)
```

```python
df_encoded.head()
```

```python
target_mapping = {
    'No_Disease': 0,
    'Low_Risk': 1,
    'Moderate_Risk': 2,
    'Severe_Disease': 3,
    'High_Risk': 4
}

df_encoded['Target'] = df_equal['Target'].map(target_mapping)
```

```python
sns.countplot(data=df_encoded,x=df_encoded['Target'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

x = df_encoded.drop('Target', axis=1)
y = df_encoded['Target']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
model_random = RandomForestClassifier(n_estimators=100, random_state=42)

model_random.fit(X_train, y_train)
model_random.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA

# pca = PCA(0.95)
# X_pca = pca.fit_transform(X_train)
# X_pca.shape
pca = PCA(0.95)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
X_train_pca.shape

```

```python
pca.explained_variance_ratio_
```

```python

model_random_pca = RandomForestClassifier(n_estimators=100, random_state=42)
model_random_pca.fit(X_train_pca, y_train)


score = model_random_pca.score(X_test_pca, y_test)
print(f"Accuracy with PCA: {score:.2f}")
```

```python
from sklearn.metrics import confusion_matrix

y_pred_pca = model_random_pca.predict(X_test_pca)


cm = confusion_matrix(y_test, y_pred_pca)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=model_random_pca.classes_,
            yticklabels=model_random_pca.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Random Forest with PCA')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_pca))
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix





X_oneHot_encoded = df_equal.drop('Target', axis=1)
y = df_equal['Target'].map({
    'No_Disease': 0,
    'Low_Risk': 1,
    'Moderate_Risk': 2,
    'Severe_Disease': 4,
    'High_Risk': 3
})


preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), object_columns)
    ],
    remainder='passthrough'
)


pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])


X_train_oneHot_encoded, X_test_oneHot_encoded, y_train, y_test = train_test_split(X_oneHot_encoded, y, test_size=0.2, random_state=30)


pipeline.fit(X_train_oneHot_encoded, y_train)


y_pred = pipeline.predict(X_test_oneHot_encoded)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    use_label_encoder=False,
    eval_metric='mlogloss'
)

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'))
])
pipeline.fit(X_train_oneHot_encoded, y_train)
y_pred = pipeline.predict(X_test_oneHot_encoded)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score



selected_features = [
    'Sugar in urine',
    'Pus cells in urine',
    'Serum phosphate level',
    'Serum albumin level',
    'White blood cell count (cells/cumm)',
    'Specific gravity of urine',
    'Serum creatinine (mg/dl)',
    'Duration of diabetes mellitus (years)',
    'Cystatin C level',
    'Estimated Glomerular Filtration Rate (eGFR)',
    'Random blood glucose level (mg/dl)',
    'Sodium level (mEq/L)',
    'Pedal edema (yes/no)'
]


X = df_equal[selected_features].copy()
y = df_equal['Target']


X['Sugar in urine'] = X['Sugar in urine'].map({'normal': 0, 'abnormal': 1})
X['Pus cells in urine'] = X['Pus cells in urine'].map({'normal': 0, 'abnormal': 1})
X['Pedal edema (yes/no)'] = X['Pedal edema (yes/no)'].map({'no': 0, 'yes': 1})


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)


rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)


y_pred = rf_model.predict(X_test)


print("Random Forest Results (Selected Features + Standard Scaled):")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```
